## Question 1.1. On considère la fonction $f(x) = \dfrac{x^2+3x+7}{x+2}$

**1.** Représenter cette fonction sur $[-5,5]$ dans un repère $[-20,50]$ en abscisses et $[-15,15]$ en ordonnée. On mettra les grilles sur la figure.

```python
import matplotlib.pyplot as plt
import numpy as np

def f(x):
    return (x**2 + 3*x + 7) / (x + 2)

x = np.linspace(-5, 5, 1000)
y = f(x)

plt.plot(x, y, label="f(x)")
plt.plot([-2, -2], [-15, 15],color='r', linestyle="--", label="Asymptote verticale")

plt.xlabel("x")
plt.ylabel("y")
plt.title("Graphique de la fonction f(x)")
plt.xlim(-5, 5)
plt.ylim(-15, 15)
plt.grid() 
plt.legend()

plt.show()
```

**2.** Justifier que la fonction $f$ admet un minimum compris entre 0 et 50


#Justifier que la fonction $f$ admet un minimum compris entre 0 et 50

Pour justifier que la fonction $f$ admet un minimum compris entre 0 et 50,ont peut etudier le comportement de la $f$ aux limites à droite et à gauche de l'asymptote verticale en $x=-2$.

Tout d'abord, en se rapprochant de l'asymptote verticale par la gauche, c'est-à-dire pour $x$ tendant vers $-2^-$, la fonction $f$ devient de plus en plus négative. En effet, le numérateur de la fraction est une fonction quadratique dont le coefficient dominant est positif, tandis que le dénominateur est négatif. Ainsi, le quotient $(x^2 + 3x + 7)/(x+2)$ tend vers $-\infty$ lorsque $x$ tend vers $-2^-$.

Ensuite, en se rapprochant de l'asymptote verticale par la droite, c'est-à-dire pour $x$ tendant vers $-2^+$, la fonction $f$ devient de plus en plus positive. En effet, le numérateur est toujours positif tandis que le dénominateur tend vers $0$ par valeurs positives, ce qui implique que le quotient tend vers $+\infty$ lorsque $x$ tend vers $-2^+$.

On peut donc en déduire que la fonction $f$ admet un minimum entre $-2$ et $+\infty$ car la fonction est continue sur cet intervalle. Par ailleurs, on peut remarquer que le minimum est atteint pour une valeur de $x$ positive car la fonction tend vers $+\infty$ lorsque $x$ tend vers $-\infty$. Enfin, en observant la courbe, on peut conjecturer que le minimum se situe entre $0$ et $5$.


**3.**  En utilisant une descente de gradient avec un pas $\alpha = 0.01$, trouver une valeur approchée de ce minumum. On initialisera la valeur de $x_0$ à $50$,
et on ferra 10000 itérations.

```python
def gradient_descent(alpha, x0, iterations):
    x = x0
    history = [x]
    for i in range(iterations):
        grad = (f(x + 0.0001) - f(x)) / 0.0001
        x = x - alpha * grad
        history.append(x)
    return x, history

alpha = 0.01
x0 = 50
iterations = 10000
x_min, history = gradient_descent(alpha, x0,iterations)

print("Valeur approchée du minimum :", x_min)

x = np.linspace(-5, 5, 1000)
y = f(x)
plt.plot(x, y, label="f(x)")
plt.plot([-2, -2], [-15, 15], linestyle="--", label="Asymptote verticale")
plt.plot(x0, f(x0), marker="o", label="Initialisation")
plt.plot(history, [f(x) for x in history], marker="o", label="Descente de gradient")
plt.xlabel("x")
plt.ylabel("y")
plt.title("Graphique de la fonction f(x)")
plt.xlim(-5, 5)
plt.ylim(-15, 15)
plt.grid()
plt.legend()
plt.show()
```

## Question 1.2. On considère la fonction $g(x)=x+e^{x}$


**1.** En représentant graphiquement le fonction $g$, montrer que l'équation $g(x)=0$ admet une solution unique et trouver un intervalle dans lequel cette solution se trouve.

```python
def g(x):
    return(x+np.exp(x))

x = np.linspace(-5, 5, 1000)
y = g(x)

plt.plot(x, y, label="f(x)")

plt.xlabel("x")
plt.ylabel("y")
plt.title("Graphique de la fonction g(x)")
plt.xlim(-5, 5)
plt.ylim(-15, 15)
plt.grid() 
plt.legend()

plt.show()
```

**2.** En utilisant la méthode de la dichotomie que vous implémenterez, trouver une valeur approchée de cette solution à $10^{-10}$ près

```python
def dichotomie():
    a = -1.0
    b = 0.0
    while b - a > 1e-10:
        m = (a + b) / 2.0
        if g(m) * g(a) < 0:
            b = m
        else:
            a = m
    return (a + b) / 2.0

solution = dichotomie()
print(f"La solution de g(x) = 0 est {solution}")
```

**3.** Placer ce point sur le graphique.

```python
x = np.linspace(-5, 5, 1000)
y = g(x)

plt.plot(x, y, label="f(x)")
plt.plot(0, dichotomie(), marker="o", label="g(x)=0")

plt.xlabel("x")
plt.ylabel("y")
plt.title("Graphique de la fonction g(x)")
plt.xlim(-5, 5)
plt.ylim(-5,5)
plt.grid() 
plt.legend()
```

Dans cette colonne, on va tout convertir en float. En particulier, on va faire la somme des capacités de toutes les caméras du téléphone et c'est ce résultat qui sera la valeur finale. Gardez le résultat final dans la variable `data`

```python
def con_fac(st):
    st=st.replace('MP', '').strip()
    st=st.replace('D', '').strip()
    st=st.replace('ToF', '1').strip()
    cat=""
    nmb=[]
    for i in st:
        if i !='+':
            cat +=i
        else:
            nmb.append(float(cat))
            cat = ''
    nmb.append(float(cat))
    
    return sum(nmb)

phone["Camera (MP)"] = phone["Camera (MP)"].apply(con_fac)

phone
```
